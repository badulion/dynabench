```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import h5py
```

```python
f = h5py.File("../data/wave/0.hdf5")
ds = f['data']
grid = f['points']
points_full = np.reshape(grid, (-1, 2))
points_high = points_full[f['indices_high']]
points_low = points_full[f['indices_low']]
```

```python
plt.gcf().set_size_inches(6.4, 6.4)
plt.pcolor(grid[:,:,0], grid[:,:,1], ds[100, 1])
plt.scatter(points_low[:,0], points_low[:,1], color="black")
plt.axis('equal')
plt.show()
```
